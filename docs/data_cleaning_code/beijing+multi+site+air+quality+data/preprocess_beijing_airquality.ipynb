```python
import numpy as np
import pandas as pd
import os
```

```python
directory = os.listdir()[0]
```

```python
files = os.listdir(directory)
```

```python
dfs = []
for file in files:
    dfs.append(pd.read_csv(directory+'/'+file))
```

```python
means_pm25 = []
means_pm10 = []
means_so2 = []
means_no2 = []
means_co = []
means_o3 = []
means_temp = []
means_pres = []
means_dewp = []
means_rain = []
means_wspm = []

for df in dfs:
    means_pm25.append(df['PM2.5'].mean())
    means_pm10.append(df['PM10'].mean())
    means_so2.append(df['SO2'].mean())
    means_no2.append(df['NO2'].mean())
    means_co.append(df['CO'].mean())
    means_o3.append(df['O3'].mean())
    means_temp.append(df['TEMP'].mean())
    means_pres.append(df['PRES'].mean())
    means_dewp.append(df['DEWP'].mean())
    means_rain.append(df['RAIN'].mean())
    means_wspm.append(df['WSPM'].mean())
```

```python
for i in range(len(dfs)):
    dfs[i]['PM2.5'].fillna(means_pm25[i], inplace = True)
    dfs[i]['PM10'].fillna(means_pm10[i], inplace = True)
    dfs[i]['SO2'].fillna(means_so2[i], inplace = True)
    dfs[i]['NO2'].fillna(means_no2[i], inplace = True)
    dfs[i]['CO'].fillna(means_co[i], inplace = True)
    dfs[i]['O3'].fillna(means_o3[i], inplace = True)
    dfs[i]['TEMP'].fillna(means_temp[i], inplace = True)
    dfs[i]['PRES'].fillna(means_pres[i], inplace = True)
    dfs[i]['DEWP'].fillna(means_dewp[i], inplace = True)
    dfs[i]['WSPM'].fillna(means_wspm[i], inplace = True)
    dfs[i]['RAIN'].fillna(means_rain[i], inplace = True)
    dfs[i]['wd'].fillna('None', inplace = True)
    
```

```python
df_concat = pd.concat(dfs, axis = 0)
```
